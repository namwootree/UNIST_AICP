# Preprocessing

## Drop Null

```python
drop_null_df = df.dropna()
```

```python
drop_null_df
```

## Set Index for Date

```python
dynmom = drop_null_df.copy()
dynmom['DATE'] = drop_null_df['DATE'].apply(lambda x : str(x))
```

```python
dynmom['year'] = dynmom['DATE'].str[:4]
dynmom['month'] = dynmom['DATE'].str[4:6]
```

```python
dynmom.index = dynmom['DATE']
dynmom = dynmom.drop(columns='DATE')
```

```python
dynmom['year'] = dynmom['year'].apply(lambda x : int(x))
dynmom['month'] = dynmom['month'].apply(lambda x : int(x))
```

# Visualization

## Winner VS Loser

```python
def winner_vs_loser(x, winner, loser, data, ylabel, title):
  plt.figure(figsize=(15, 7))
  sns.lineplot(x=x,
              y=winner,
              data=data,
              label=winner,
              alpha=0.5)

  sns.lineplot(x=x,
              y=loser,
              data=data,
              label=loser,
              alpha=0.5)

  plt.xticks([dynmom.index[i] for i in range(0,len(dynmom.index), 12)])

  plt.tick_params(axis='x',
                  direction='out',
                  labelrotation=45,
                  length=1,
                  pad=10,
                  labelsize=5,
                  width=0.1)


  plt.title(title)
  plt.xlabel('DATE')
  plt.ylabel(ylabel)

  plt.legend(fontsize=15)

  plt.show()
```

```python
winner_vs_loser(x=dynmom.index,
                winner='winner',
                loser='loser',
                data=dynmom,
                ylabel='Winner & Loser',
                title='Winner VS Loser')
```

```python
winner_vs_loser(x=dynmom.index,
                winner='winner',
                loser='loser',
                data=dynmom,
                ylabel='Winner & Loser',
                title='Winner VS Loser')
```
